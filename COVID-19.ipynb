```python
covid2 = covid.iloc[103:110]
covid2
```

```python
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc # 한글 처리
f_path = 'C:/Windows/Fonts/malgun.ttf'
font_name=font_manager.FontProperties(fname=f_path).get_name()
rc('font', family = font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
covid2.set_index('국가',inplace=True)
```

```python
f, ax = plt.subplots(1, 3, figsize=(16,8))

covid2['테스트양'].plot.pie(explode=[0.02,0.02,0.02,0.02,0.02,0.02,0.02],ax=ax[0], autopct='%1.1f%%')
ax[0].set_title('코로나 검사')

covid2['측정양'].plot.pie(explode=[0.02,0.02,0.02,0.02,0.02,0.02,0.02], ax=ax[1], autopct='%1.1f%%')
ax[1].set_title('코로나 확진자')

covid2['코로나율(%)'].plot.pie(explode=[0.02,0.02,0.02,0.02,0.02,0.02,0.02], ax=ax[2], autopct='%1.1f%%')
ax[2].set_title('전체 코로나 %')

plt.show()
```
