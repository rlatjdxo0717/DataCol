```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
ax = plt.subplots()
```

```python
ax = sns.distplot(tips['total_bill'], rug=True)
```

```python
tips
```

```python
ax2 = plt.subplots()
```

```python
ax2 = sns.regplot(x='total_bill', y='tip', data=tips, fit_reg = False)
ax2.set_title('total Bill and tip')
ax2.set_xlabel('Total Bill')
ax2.set_ylabel('Tip')
```

```python
ax3 = sns.jointplot(x='total_bill', y='tip', data=tips)
ax3.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax3.fig.suptitle('joint Total Bill and Tip', fontsize=10)
```

```python
ax4 = sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
ax4.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax4.fig.suptitle('joint Total Bill and Tip', fontsize=10)
```

```python
ax3 = sns.jointplot(x='total_bill', y='tip', data=tips)
ax3.set_axis_labels(xlabel='Total Bill', ylabel='Tip')
ax3.fig.suptitle('joint Total Bill and Tip', fontsize=10)
```

```python
ax4 = sns.barplot(x='time', y='tip', data=tips)
ax4.set_title('total Bill and tip')
ax4.set_xlabel('Total Bill')
ax4.set_ylabel('Tip')
```

```python
ax5 = sns.boxplot(x='time', y='tip', data=tips)
ax5.set_title('total Bill and tip')
ax5.set_xlabel('Total Bill')
ax5.set_ylabel('Tip')
```

```python
ax6 = sns.violinplot(x='time', y='tip', hue='sex', data=tips)
ax6.set_title('total Bill and tip')
ax6.set_xlabel('Total Bill')
ax6.set_ylabel('Tip')
```

```python
ax7 = sns.violinplot(x='time', y='tip', hue='sex', data=tips, split=True)
ax7.set_title('total Bill and tip')
ax7.set_xlabel('Total Bill')
ax7.set_ylabel('Tip')
```
